## Data Preprocessing
### Encoding the categorical values  

### converting the categorical variables into contious variables to feed the models

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
df = df.apply(LabelEncoder().fit_transform)
df
```

```python
#checking for the unique data of categorical variables
for i in (df.columns):
    print(i,':',df[i].value_counts())
```

```python
#converting the dataset into Independent and dependent variables

X = df.drop(['class'],axis=1)
y=df['class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=24)
```

##  Model Building

### Using Decision tree model to predict the data and check accuracy score

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=0, max_depth=5)
dt.fit(X_train, y_train)
```

```python
print("The score for training dataset:",dt.score(X_train,y_train))
```

```python
pred = dt.predict(X_test)
print("THe prediction score X testing data:", pred)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, pred)
print("The accuracy score of predicted data:",acc)
```

We have got a good score from this model, We can use another model to check the score.

### Random Forest Model to compare accuracy score of previous model

```python
#Instantiating the Model
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth=5)
```

```python
#Fit the Model and check score on train dataset
rf.fit(X_train, y_train)
rf.score(X_train, y_train)
```

```python
#Checking the score for test data sets

pred = rf.predict(X_test)
rf.score(X_test, y_test)
```

```python
accuracy_score(y_test, pred)
```
